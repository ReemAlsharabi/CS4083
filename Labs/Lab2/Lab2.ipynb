# Task 5

```python
def to_lower_case(text):
  low = []
  for par in text:
    par = par.lower()
    low.append(par)
  return low

clean_text = to_lower_case(raw_data)
clean_text[0]
len(clean_text)
```

# Task 6

```python
from nltk import word_tokenize, sent_tokenize

import nltk
nltk.download('punkt')

clean_text_2=[]

clean_text_2=[word_tokenize(par) for par in clean_text]
```

```python
clean_text_2[4]
```

# Task 7

```python
import re

clean_text_3=[]

for par in clean_text_2:
    clean=[]
    for w in par:
        res=re.sub(r'[^\w\s\-\_]','',w)
        if res != "":
            clean.append(res)
    clean_text_3.append(clean)

clean_text_3[98]
```

## Task 8
Stopwords elimination

```python
from nltk.corpus import stopwords

clean_text_4=[]
nltk.download('stopwords')

for par in clean_text_3:
    clean=[]
    for w in par:
        if not w in stopwords.words('english'):
            clean.append(w)
    clean_text_4.append(clean)

clean_text_4[94]
```

## Task 9
Stemming

```python
from nltk.stem.porter import PorterStemmer
port=PorterStemmer()

clean_text_5=[]

for par in clean_text_4:
  clean = [port.stem(w) for w in par]
  clean_text_5.append(clean)

clean_text_5[94]
```

# Task 9
lemmatization

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnet = WordNetLemmatizer()
import nltk
nltk.download('wordnet')
clean_text_6 = []

for par in clean_text_5:
  clean = [wnet.lemmatize(w) for w in par]
  clean_text_6.append(clean)

clean_text_6[94]
```
